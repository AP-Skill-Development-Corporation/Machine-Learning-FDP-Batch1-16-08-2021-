#### Replacing missing values
- If the data is numerical, replace Null values with either with 0 or mean() or any other value
- If the data is objective, replace Null values with either with most occured name or any other object.
    - fillna()

```python
data2 = df4.copy()
```

```python
data2.head()
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

```python
data2.info()
```

```python
m = data2['Product_Base_Margin'].mean()
m
```

```python
data2['Product_Base_Margin'] = data2['Product_Base_Margin'].fillna(m)
```

```python
data2.shape
```

```python
data2.isnull().sum()
```
